```python
def convert(string):
    li = list(string.split('  '))
    return li
```

```python
coursera['Skills'] = coursera['Skills'].apply(convert)
```

```python
coursera.head()
```

```python
coursera['Course Description'][1260]
```

```python
coursera['Course Description'] = coursera['Course Description'].apply(lambda x:x.split())
```

```python
coursera.head()
```

```python
coursera['University'] = coursera['University'].apply(convert)
```

```python
coursera.head()
```

```python
coursera['University'] = coursera['University'].apply(lambda x:[i.replace(" ","") for i in x])
coursera['Skills'] = coursera['Skills'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
coursera['tags'] = coursera['University'] + coursera['Skills'] + coursera['Course Description']
```

```python
coursera.head()
```

```python
new_df = coursera[['Course Name', 'Course Rating', 'Course URL', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.rename(columns = {'Course Name':'course_name'}, inplace = True)
new_df.rename(columns = {'Course Rating':'course_rating'}, inplace = True)
new_df.rename(columns = {'Course URL':'course_url'}, inplace = True)
```

```python
new_df.head(-10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(course):
    course_index = new_df[new_df['course_name'] == course].index[0]
    distances = similarity[course_index]
    
    course_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:11]
    
    for i in course_list:
        print(new_df.iloc[i[0]].course_name)
```

```python
recommend('Pixel Art for Video Games')
```
